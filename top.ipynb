```python
from pprint import pprint
import requests
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', 1000)
from wordcloud import WordCloud
from matplotlib import pyplot as plt
```

```python
params = {
    "per-page": 200,
    "mailto": "nick.young@auckland.ac.nz"
}
response = requests.get("https://api.openalex.org/authors?filter=affiliations.institution.country_code:nz&sort=cited_by_count:desc", params=params).json()
df = pd.json_normalize(response["results"])
```

```python
df["affiliations"] = df.affiliations.apply(lambda a: [i["institution"]["display_name"] + " " + str(i["years"]) for i in a if i["institution"]["country_code"] == "NZ"])
```

```python
df["topic"] = df.topics.apply(lambda x: x[0]["domain"]["display_name"] if x else None)
```
